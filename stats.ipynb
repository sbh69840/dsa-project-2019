```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pandas as pd
stats = pd.read_csv("SEARCH.csv")
stats.head(20)
```

```python
plt.plot(stats["No. Elements"],stats["CPU brute force"],"-",color="b",label="CPU brute force")
plt.legend()
plt.savefig("brute.png")
```

```python
plt.plot(stats["No. Elements"],stats["CPU iterative"],"-",color="r",label="CPU iterative segment tree")
plt.legend();
plt.savefig("iterative.png")
```

```python
plt.plot(stats["No. Elements"],stats["CPU recursive"],"-",color="r",label="CPU recursive segment tree")
plt.legend();
plt.savefig("recur.png")
```

```python
plt.plot(stats["No. Elements"],stats["CPU brute force"],"-",color="g",label="CPU brute force")
plt.plot(stats["No. Elements"],stats["CPU iterative"],"-",color="b",label="CPU iterative segment tree")
plt.plot(stats["No. Elements"],stats["CPU recursive"],"-",color="r",label="CPU recursive segment tree")
plt.legend()
plt.savefig("all.png")
```
